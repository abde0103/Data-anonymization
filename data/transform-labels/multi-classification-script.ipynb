each label will be represented by a one hot vector which is essential for our network layer afterwards.


For example 'PER' will be represented as [0,1,0,0,0] 

```python
# shape of new_df is (5000,5) pour 
new_df = pd.DataFrame(np.zeros((df.shape[0],5)), dtype=int)
```

```python
new_df.head()
```

```python
labels = {'PER':1,'MISC':2,'LOC':3,'ORG':4,'O':0}
```

```python
labels.keys()
```

```python
def transform(ligne):
    for key in labels.keys():
        if key in df[0][ligne]:
            new_df.iloc[ligne,labels[key]]=1
            break;
```

```python
# fill in the new_df
```

```python
for ligne in range(df.shape[0]):
    transform (ligne)
```
